```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import BernoulliRBM
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
iris = load_iris()
X, y = iris.data, iris.target
```

```python
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=(0.2), random_state=34)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_tr)
X_test = scaler.transform(X_te)
```

```python
rbm = BernoulliRBM(n_components=2, learning_rate=0.01, n_iter=100)
```

```python
X_train_transformed = rbm.fit_transform(X_train)
X_test_transformed = rbm.transform(X_test)
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1,2,1)
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_te, cmap=plt.cm.Set1)
plt.title("Original Data")

plt.subplot(1,2,2)
plt.scatter(X_test_transformed[:, 0], X_test_transformed[:, 1], c=y_te, cmap=plt.cm.Set1 )
plt.title("After RBM Transformation")

plt.show()
```
